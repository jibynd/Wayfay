```python
scores = {"mse":[]}
start = time.time(); n_iter = 100
for i in tqdm(range(n_iter)):
    data = ExampleSelectOne(n = 60, sig=1, with_test = True, seed = i + 1).data
    #data.train.to_df(); data.test.to_df()
    yt = data.test.y; X_ = data.test.X 
    model = bayes.Wayfay(nIter=100, ld0 = 1, d0 = 1, eps=1/10000)
    model.fit(data.train.y, data.train.X) 
    md = st.mode(model.gammas.storage.sum(1))
    gam = model.gammas.storage.mean(axis=0)  #> 0.5
    argz = np.argsort(gam)[-int(md):]
    gam = np.zeros_like(gam)
    gam[argz] = 1; #print(gam, md)
    tg = data.betas > 0; 
    sc = (gam == tg).all() 
    scores["mse"] += [sc]
    #scores["corr"] += [np.corrcoef(preds, yt)[0,1]]
end = time.time() - start 
df = pd.DataFrame(scores)
print(df)
#res_df.to_csv("Data/Ex3_100_1_bis.csv")

```
